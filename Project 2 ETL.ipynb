```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
car_file = "Resources/Car_sales.csv"
```

```python
car_df = pd.read_csv(car_file)
car_df.head()
```

```python
details_df = car_df[['Manufacturer', 'Model', 'Vehicle type', 'Engine size', 'Horsepower','Wheelbase', 'Width', 'Length','Curb weight', 'Fuel capacity','Fuel efficiency','Latest Launch']]
details_df.head()
```

```python
price_df = car_df[['Manufacturer', 'Model', '4-year resale value', 'Price in thousands']]
price_df.head()
```

```python
manufacturer_sales = car_df[['Manufacturer', 'Sales in thousands','Price in thousands']]
manufacturer_sales['Price in thousands']
manufacturer_sales.head()
```

```python
manufacturer_sales.dtypes
```

```python
group = manufacturer_sales.groupby('Manufacturer')
```

```python
price = group['Price in thousands'].mean()
```

```python
sales = group['Sales in thousands'].sum()
```

```python
sales_df = pd.DataFrame(sales)
price_df = pd.DataFrame(price)
price_df.head()
```

```python
manufacturer_input=sales_df.merge(price_df, on='Manufacturer')

manufacturer_input.head()
```
