```python
import pandas as pd
import sys
import pandas as pd
import pymongo
import json
from sqlalchemy import create_engine
```

```python
car_file = "Resources/Car_sales.csv"
```

```python
car_df = pd.read_csv(car_file)
car_df.head()
```

```python
car_df['Manufacturer']=car_df['Manufacturer'].str.strip()
car_df['Model']=car_df['Model'].str.strip()
car_df.head()
```

```python
details_df = car_df[['Manufacturer', 'Model', 'Vehicle type', 'Engine size', 'Horsepower','Wheelbase', 'Width', 'Length','Curb weight', 'Fuel capacity','Fuel efficiency','Latest Launch']]
details_df.head()
```

```python
price_data = car_df[['Manufacturer', 'Model', '4-year resale value', 'Price in thousands', 'Sales in thousands']]
price_data.head()
```

```python
manufacturer_sales = car_df[['Manufacturer', 'Sales in thousands','Price in thousands']]
manufacturer_sales['Price in thousands']
manufacturer_sales.head()
```

```python
manufacturer_sales.dtypes
```

```python
group = manufacturer_sales.groupby('Manufacturer')
```

```python
price = group['Price in thousands'].mean()
```

```python
sales = group['Sales in thousands'].sum()
```

```python
sales_df = pd.DataFrame(sales)
price_df = pd.DataFrame(price)
price_df.head()
```

```python
sort_price = price_df.sort_values(by = 'Price in thousands')
sort_price
sort_sales = sales_df.sort_values(by = 'Sales in thousands')
sort_sales.head()
```

```python
manufacturer_input=sales_df.merge(price_df, on='Manufacturer')

manufacturer_input.head()
```

```python
manufacturer_input_json = manufacturer_input.to_json(orient='index')
print(manufacturer_input_json)
```

```python
details_json = details_df.to_json(orient='records')
print(details_json)
```

```python
price_json = price_data.to_json(orient='records')
print(price_json)
```

```python
mng_client = pymongo.MongoClient('localhost', 27017)
mng_db = mng_client['ETL_project'] 
db_cm = mng_db['car_price' ]

db_cm.remove()
db_cm.insert(json.loads(price_json))
```

```python
db_cm = mng_db['car_details' ]

db_cm.remove()
db_cm.insert(json.loads(details_json))
```

```python
db_cm = mng_db['manufacturer_aggregate' ]

db_cm.remove()
db_cm.insert(json.loads(manufacturer_input_json))
```
